# Filling the Null entries by inputting the missing data by the Average/Mean of the Particular feature:

```python
df['education'].fillna(np.mean(df['education']),inplace=True)
df['glucose'].fillna(np.mean(df['glucose']),inplace=True)
df['BMI'].fillna(np.mean(df['BMI']),inplace=True)
df['totChol'].fillna(np.mean(df['totChol']),inplace=True)
df['cigsPerDay'].fillna(np.mean(df['cigsPerDay']),inplace=True)
df['BPMeds'].fillna(np.mean(df['BPMeds']),inplace=True)
df['heartRate'].fillna(np.mean(df['heartRate']),inplace=True)
```

# PREPARING THE DATA FOR TRAINING AFTER VISUALISING:

```python
#splitting data to X and Y:
X = df.drop('TenYearCHD',axis=1)
Y = df['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.3,shuffle = True)
```

# MODEL SELECTION AND FITTING TRAINING DATA:

# 1)LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

# Scaling the Data for Better performance:

```python
from sklearn import preprocessing
```

```python
scaler = preprocessing.StandardScaler().fit(xtrain)
```

```python
Xscaled = scaler.transform(xtrain)
```

```python
#Fitting the data:
lr.fit(Xscaled,ytrain)
```

# MODEL EVALUATION AND CROSS-VALIDATION:

```python
#USING R-SQUARE METRIC FOR MODEL EVALIUATION ON TRAINING DATA:
lr.score(Xscaled,ytrain)
```

```python
k = preprocessing.StandardScaler().fit(xtest)
```

```python
scaledtest = k.transform(xtest)
```

```python
#USING R-SQUARE METRIC FOR MODEL EVALIUATION ON TESTING DATA:
lr.score(scaledtest,ytest)
```

# 2)DECISION TREE CLASSIFIER:

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTc = DecisionTreeClassifier(random_state = 1,max_depth = 3)
```

```python
#Fitting the training data:
DTc.fit(Xscaled,ytrain)
```

```python
#USING R-SQUARE METRIC FOR MODEL EVALIUATION ON TRAINING DATA:
DTc.score(Xscaled,ytrain)
```

```python
#USING R-SQUARE METRIC FOR MODEL EVALIUATION ON TESTING DATA:
DTc.score(scaledtest,ytest)
```

# 3)RANDOM FORTEST CLASSIFER

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFc = RandomForestClassifier(max_depth = 3,random_state = 1)
```

```python
#Fitting the training data: 
RFc.fit(Xscaled,ytrain)
```

```python
#USING R-SQUARE METRIC FOR MODEL EVALIUATION ON TRAINING DATA:
RFc.score(Xscaled,ytrain)
```

```python
#USING R-SQUARE METRIC FOR MODEL EVALIUATION ON TESTING DATA:
RFc.score(scaledtest,ytest)
```
